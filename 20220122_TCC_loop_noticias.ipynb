#Obtendo o conjunto de notícias de uma página de um site

```python
# Script usando BeautifulSoup
#TCC - 21/01/2022
#Inserido comandos para identificar e armazenar o link da notícia

from urllib.request import urlopen                # Biblioteca que permite fazer a solicitação de acesso de site(HTTP) e obtém a resposta
from bs4 import BeautifulSoup                     # Transforma a resposta obtida do url open em um objeto navegável de modo mais simples
             
url = 'https://www.gov.br/mme/pt-br/assuntos/noticias'    #variável que armazena o site a ser acessado

response = urlopen(url)                          # Abre o site desejado e armazena
html = response.read()                           # Lê o conteúdo da resposta do servidor
#html                                            #Usar este comando assim vai trazer a programação em HTML desestruturada

#----------------- 1 -Tratamento dos dados-------------------

#---1.1 - Convertendo de bytes para string---

#print(type(html))                               #Informa o tipo do dado antes do tratamento - byte
html = html.decode('utf-8')
#type(html)                                       #informa o tipo do dado após o tratamento - vira string por causa da codificação


#----------------1.2 - Eliminando caracteres indesejados e espaços em branco-------------

html.split()                                    #Eliminando caracteres de tabulação e quebra de linha - separa por espaço entre caracteres
" ".join(html.split())                          #Eliminando espaço em branco - o dado precisa estar convertido em string
#htmL

" ".join(html.split()).replace('> <', '><')     #Eliminando espaço em branco entre as tags > < 
#html
```

```python
#------- 2-Criação do objeto Beautifulsoup--------

soup = BeautifulSoup(html, 'html.parser')        #Criação do objeto Beautifulsoup 
#soup                                            #Permite a visualização do objeto
#soup.html                                       #acessando as tags do objeto soup com o tratamento de dados da veriável html
#soup.title                                       #título entre tags HTML de título
#soup.title.get_text()                            #título sem as tags de título
#print("Título da página: ")
#print(soup.title.get_text())                    #Parei na análise do item 5
```

#Apresentando a data com e sem tag

```python
#Esta rotina armazena e apresenta a data com tags e sem tags
#Data - tag +data
datacomtag = soup.find('span', {'class':'data'})
print('\n')
print('Data com tag:')
print (datacomtag,'\n')

datasemtag = soup.find('span', {'class':'data'}).get_text()
print('Data sem tag:','\n')
print (datasemtag)
```
